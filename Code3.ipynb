### Write a Python script that trains a simple linear regression model using scikit-learn. Use a dataset of your choice, split it into training and testing sets, and evaluate the model's performance.

```python
import numpy as np
import pandas as pd
dataset=pd.read_csv("weight-height.csv")
print("first five data: ")
print(dataset.head())
print("last five data: ")
print(dataset.tail())
print("Shape of the dataset: ")
print(dataset.shape)
print("check if null values are present or not")
print(dataset.isnull().sum())
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
dataset['Gender']=labelencoder.fit_transform(dataset['Gender'])
print(dataset.head())
print(dataset.tail())
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
print("Independent data: ")
print(x)
print("Dependent data: ")
print(y)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print("x_train: ", x_train)
print("x_test: ",x_test)
print("y_train: ",y_train)
print("y_test: ",y_test)
from sklearn.preprocessing import StandardScaler
standardscaler=StandardScaler()
x_train=standardscaler.fit_transform(x_train)
x_test=standardscaler.transform(x_test)
print("After scaling features: ")
print("x_train: ",x_train)
print("x_test: ",x_test)
from sklearn.linear_model import LinearRegression
linearregression=LinearRegression()
linearregression.fit(x_train,y_train)
ypred=linearregression.predict(x_test)
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
print("Mean_squared_error: ",mean_squared_error(y_test,ypred))
print("mean_Absolute_error: ",mean_absolute_error(y_test,ypred))
print("r2_score: ",r2_score(y_test,ypred))
```
